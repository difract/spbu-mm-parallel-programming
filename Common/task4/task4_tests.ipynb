```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
def plotting(datatable):
    boxprops = dict(color='k')
    plot = datatable.boxplot(by = 'label',column=['elapsed'], showfliers=False, boxprops=boxprops)
    plot.set_ylabel('average time(ms)')
    plot.set_xlabel('')
    plot.get_figure().suptitle("")
    return plot
```

```python
def single_plotting(datatable):
    boxprops = dict(color='k')
    plot = datatable.boxplot(column=['elapsed'], showfliers=False, boxprops=boxprops)
    plot.set_ylabel(datatable['label'][0])
    plot.set_xlabel('')
    plot.get_figure().suptitle("")
    return plot
```

```python
df = pd.read_csv("lazy_all_jmeter_results200x200.csv")
plot = plotting(df)
plot.set_title('200 Threads are doing 3 operations on lazyset')
```

```python
df = pd.read_csv("lazy_all_jmeter_results1900x100.csv")
plot = plotting(df)
plot.set_title('5000 Threads are doing 3 operations on lazyset')
```
